# Pyplot Tutorial

matplotlib.pyplot is a collection of command style func similar to matlab

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.xlabel("indices")
plt.ylabel("nums")
plt.title("blabla")
plt.grid() #grid on
plt.show()
```

if you provide a single list or array, the indices are x axis

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel("indices")
plt.ylabel("nums")
plt.title("blabla")
plt.grid() #grid on
plt.show()
```

```python
import numpy as np
t = np.arange(0, 10, 1)

plt.plot(t, t**2, "bs--", label = "^2")
plt.plot(t, t**2.2, "ro--", label = "^2.2")
plt.plot(t, t**2.5, "g^-", label = "^2.5")
plt.grid()
plt.legend()
plt.show()
```

#### Plot multiple figures

```python
def f(t):
        return np.exp(-t) + np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure(1)
#the subplot command specifies numrows and numcols, fignum where fignum ranges from 1 to numrows*numcols

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), "b-", linewidth = 2.0)
plt.show()

plt.subplot(212)
#plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), "r--", linewidth = 2.0)
plt.show()
```

```python
plt.figure(1)
#the subplot command specifies numrows and numcols, fignum where fignum ranges from 1 to numrows*numcols

plt.subplot(121)
plt.grid()
plt.plot(t1, f(t1), "b-", linewidth = 2.0)
plt.show()

plt.subplot(122)
#plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), "r--", linewidth = 2.0)



plt.figure(2)
plt.plot(t1, np.sin(2*np.pi*t1), "g+-", linewidth = 2.0)
plt.show()
```
